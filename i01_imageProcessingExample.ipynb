# Basic image processing examples

Here are a few basic examples of image processing with spline, convolved, and guassian type filters for edge, shift, and blurred effects/processing.

```python
#===============================
%matplotlib inline

import numpy as np
from scipy import signal, misc
import matplotlib.pyplot as plt
```

### Original image

```python
image = misc.ascent().astype(np.float32)

plt.imshow(image)
plt.gray()
plt.title('Original')
plt.show()
```

### spline filter (edge detection) of original image

```python
derfilt = np.array([1.0, -2, 1.0], dtype=np.float32)
ck = signal.cspline2d(image, 8.0)
new_image = (signal.sepfir2d(ck, derfilt, [1]) + signal.sepfir2d(ck, [1], derfilt))

plt.imshow(new_image)
plt.gray()
plt.title('Spline (edge filter)')
plt.show()
```

### convolution  (shift-invariance) of original image

```python
image = misc.ascent()

w = np.zeros((50, 50))
w[0][0] = 2.0
w[49][25] = 1.2
image_new = signal.fftconvolve(image, w)

plt.imshow(image_new)
plt.gray()
plt.title('fft-convolved (shift)')
```

### Guassian filtering (blurring) of original image

```python
image = misc.ascent()

w = signal.gaussian(50, 10.0)
image_new = signal.sepfir2d(image, w, w)

plt.imshow(image_new)
plt.gray()
plt.title('guass-filtered (blurred)')
```
